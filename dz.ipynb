```python
import matplotlib.pyplot as plt
```

```python
df_house_data = pd.read_csv("kc-house-data (1).csv")
waterfront_counts = df_house_data['waterfront'].value_counts()
plt.figure(figsize=(2, 2))
plt.pie(waterfront_counts, labels=['Нет', 'Да'], autopct='%1.1f%%', startangle=90)
plt.title('Распределение домов от наличия вида на набережную')
plt.axis('equal') 
plt.show()
```

```python
df_house_data['floors'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.title('Распределение этажей домов')
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(2, 3))
sns.countplot(data=df_house_data, x='condition')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов')
plt.title('Распределение состояния домов')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_house_data, x='condition', hue='floors')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов')
plt.title('Многорядная столбчатая диаграмма для состояния и этажей')
plt.legend(title='Количество этажей')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_house_data, x='lat', y='price', alpha=0.7)
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Зависимость стоимости от квадратуры жилой площади')
plt.show()
```

Выводы и наблюдения: Из графика видно, что есть положительная корреляция между квадратурой жилой площади и стоимостью недвижимости - чем больше площадь дома, тем выше его стоимость.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_house_data, x='bedrooms', y='price')
plt.xlabel('Количество комнат')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Распределение стоимости недвижимости по количеству комнат')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_house_data, x='waterfront', y='price', hue='floors')
plt.xlabel('Вид на набережную')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Распределение стоимости недвижимости по виду на набережную и этажам')
plt.legend(title='Количество этажей')
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.boxplot(data=df_house_data, x='yr_built', y='price')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Распределение стоимости недвижимости по году постройки')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.lineplot(data=df_house_data, x='yr_built', y='price')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Распределение стоимости недвижимости по году постройки')
```
